```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from astropy import units as u
from astropy import constants as const
from scipy.optimize import curve_fit
from scipy import stats
from scipy.stats import linregress
```

```python
Apollo_11 = pd.read_csv('10072.csv')
Apollo_11
```

```python
Apollo_12 = pd.read_csv('12031.csv')
Apollo_12
```

```python
#Plotting up Apollo 11
fig,ax = plt.subplots(1,1)
fig.set_size_inches(11,8.5)
fig.tight_layout()

ax.set_xlabel("Sm147")
ax.set_ylabel("Nd143")

ax.plot(Apollo_11['Sm147'], Apollo_11['Nd143'],
       color = "g",
       marker = "*",
       linestyle = "None",
       Markersize = 8);
```

```python
#Plotting up Apollo 12
fig,ax = plt.subplots(1,1)
fig.set_size_inches(11,8.5)
fig.tight_layout()

ax.set_xlabel("Rb87")
ax.set_ylabel("Sr87")

ax.plot(Apollo_12['Rb87'], Apollo_12['Sr87'],
       color = "r",
       marker = "*",
       linestyle = "None",
       Markersize = 8);
```

```python
my_fit_11 = np.polyfit(Apollo_11['Sm147'], Apollo_11['Nd143'], 4)

my_fit_11

my_fit_12 = np.polyfit(Apollo_12['Rb87'], Apollo_12['Sr87'], 4)

my_fit_12
```

```python
fitted_poly_11 = np.poly1d(my_fit_11)
fitted_poly_12 = np.poly1d(my_fit_12)
```

```python
Apollo_11['Fit'] = fitted_poly_11(Apollo_11['Sm147'])
Apollo_12['Fit'] = fitted_poly_12(Apollo_12['Rb87'])
```

```python
#Fitting Apollo 11
fig,ax = plt.subplots(1,1)
fig.set_size_inches(11,8.5)
fig.tight_layout()

ax.set_xlabel("Sm147")
ax.set_ylabel("Nd143")

ax.plot(Apollo_11['Sm147'], Apollo_11['Nd143'],
       color = "g",
       marker = "*",
       linestyle = "None",
       Markersize = 8,
       label = "Data");

ax.plot(Apollo_11['Sm147'], Apollo_11['Fit'],
       marker = "None",
       linewidth = 3,
       color = 'y',
       linestyle = "--",
       label = "Fit to Data");

ax.legend(loc=0);
```

```python
#Slope of 11
age_11 = linregress(Apollo_11['Sm147'], Apollo_11['Nd143'])
age_11
```

### Indicates that the age of sample 10072 is about 3.6 Billion years old.

```python
#Fitting Apollo 12
fig,ax = plt.subplots(1,1)
fig.set_size_inches(11,8.5)
fig.tight_layout()

ax.set_xlabel("Rb87")
ax.set_ylabel("Sr87")

ax.plot(Apollo_12['Rb87'], Apollo_12['Sr87'],
       color = "r",
       marker = "*",
       linestyle = "None",
       Markersize = 8,
       label = "Data");

ax.plot(Apollo_12['Rb87'], Apollo_12['Fit'],
       marker = "None",
       linewidth = 3,
       color = 'b',
       linestyle = "--",
       label = "Fit to Data");

ax.legend(loc=0);
```

```python
#Slope of 12
age_12 = linregress(Apollo_12['Rb87'], Apollo_12['Sr87'])
age_12
```

### The age of sample 12031 is 3.23 ± 0.07 Billion years.

From the above fitting and linear regression we see that sample 10072 collected on Apollo 11 is ~0.4 Billion years older than sample 12031 collected on Apollo 12.
